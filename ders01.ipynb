```python
def harf_notu(puan):
    if puan < 0 or puan > 100:
        return "Geçersiz"
    if puan >= 90: return "AA"
    if puan >= 85: return "BA"
    if puan >= 80: return "BB"
    if puan >= 75: return "CB"
    if puan >= 70: return "CC"
    if puan >= 65: return "DC"
    if puan >= 60: return "DD"
    return "FF"

print(harf_notu(87), harf_notu(62))
```

```python
def hesap(a, b, op):
    if op == "+": return a + b
    if op == "-": return a - b
    if op == "*": return a * b
    if op == "/":
        if b == 0:
            return None  # veya raise ValueError
        return a / b
    if op == "//":
        if b == 0:
            return None  # veya raise ValueError
        return a // b
    return None

print(hesap(10, 3, "/"), hesap(10, 0, "/"),hesap(5,2,"/"),hesap(5,2,"//"))
```

```python
def liste_istatistik(sayilar):
    if not sayilar:
        return None
    n = len(sayilar)
    toplam = sum(sayilar)
    sirali = sorted(sayilar)
    medyan = sirali[n // 2] if n % 2 else (sirali[n//2 - 1] + sirali[n//2]) / 2
    return {
        "min": min(sayilar),
        "max": max(sayilar),
        "ortalama": toplam / n,
        "medyan": medyan
    }

print(liste_istatistik([3, 1, 4, 1, 5]))
```

Regex

1) re.match(pattern, text)

Sadece baştan eşleşirse çalışır.

2) re.search(pattern, text)

Metnin her yerinde arar, ilk bulduğunu döner.

3) re.findall(pattern, text)

Bulduğu tüm eşleşmeleri liste olarak döner.

4) re.finditer(pattern, text)

Bulduklarını iterator olarak verir (match objeleri).

5) re.sub(pattern, repl, text)

Bulduklarını değiştirir.

re.compile() neden var?

Aynı pattern’i çok kullanacaksan performans ve temizlik için:

Karakter sınıfları

\d rakam, \D rakam değil

\w harf/rakam/_ , \W değil

\s boşluk/tab/newline, \S değil

Tekrarlayıcılar

+ 1 veya daha fazla

* 0 veya daha fazla

? 0 veya 1

{n} tam n kez

{n,} n veya daha fazla

{n,m} n ile m arası

Çapalar (konum)

^ satır başı

$ satır sonu

\b kelime sınırı

Gruplama

( ... ) capturing group

(?: ... ) non-capturing group



```python
import re
print(bool(re.match(r"\d+", "123abc")))   # True
print(bool(re.match(r"\d+", "abc123")))   # False
```

```python
m = re.search(r"\d+", "abc123def")
print(m.group())  # 123
```

```python
print(re.findall(r"\d+", "a1 b22 c333"))  # ['1', '22', '333']
```

```python
for m in re.finditer(r"\d+", "a1 b22 c333"):
    print(m.group(), m.start(), m.end())
```

```python
print(re.sub(r"\d+", "#", "id=123; id=456"))  # id=#; id=#
```

```python
p = re.compile(r"\b\w+\b")
print(p.findall("Merhaba dünya!"))  # ['Merhaba', 'dünya']
```

1) Greedy (açgözlü) eşleşme

.* mümkün olan en uzun eşleşmeyi alır. “Az alsın” istiyorsan .*?

```python
text = "<b>one</b><b>two</b>"
print(re.findall(r"<b>.*</b>", text))    # ['<b>one</b><b>two</b>']
print(re.findall(r"<b>.*?</b>", text))   # ['<b>one</b>', '<b>two</b>']
```

```python
text = "E-posta: ali@test.com ve veli@site.org"

pattern = r"[\w.]+@[\w.]+"
print("findall:", re.findall(pattern, text))
print("search:", re.search(pattern, text).group() if re.search(pattern, text) else None)
```

```python
text2 = """
Kişiler:
Ahmet: 0532 123 45 67
Ayşe: +90 532 987 65 43
Mehmet: 5321234567  (bu formatı yakalamayabilir, çünkü boşluk yok)
Yanlış: 0212 444 44 44 (5 ile başlamıyor)
"""

phones = re.findall(r"(?:\+90|0)?\s*5\d{2}\s*\d{3}\s*\d{2}\s*\d{2}", text2)
print("2) phones:", phones)
```

Lambda genelde map/filter/sorted gibi yerlerde “tek kullanımlık” fonksiyon için.

map() — “her elemana uygula”

Bir liste içindeki her elemana aynı işlemi uygular.

```python
sayilar = [1, 2, 3, 4]
sonuc=list(map(lambda x: x**2, sayilar))
print(sonuc)
```

```python
veri = ["10", "20", "30"]
sayilar = list(map(int, veri))
print(sayilar)  # [10, 20, 30]
```

filter() — “şartı sağlayanları seç”

Listeyi bir koşula göre süzmek.

```python
sayilar = [1, 2, 3, 4, 5, 6]
ciftler = list(filter(lambda x: x % 2 == 0, sayilar))
print(ciftler)
```

```python
sayilar = [1, 2, 3, 4, 5]
buyuk = list(filter(lambda x: x > 3, sayilar))
print(buyuk)  # [4, 5]
```

reduce() — “hepsini tek değere indir”

reduce Python’da direkt gelmez, functools içindedir.

```python
from functools import reduce
sayilar = [1, 2, 3, 4]
toplam = reduce(lambda acc, x: acc + x, sayilar, 0)
print(toplam)
```

```python
from functools import reduce

sayilar = [10, 3, 25, 7]
en_buyuk = reduce(lambda acc, x: acc if acc > x else x, sayilar)
print(en_buyuk)  # 25
```

“Pipeline” örneği (filter + map + reduce birlikte)

```python

nums = list(range(1, 11))

ciftler = filter(lambda x: x % 2 == 0, nums)
kareler = map(lambda x: x * x, ciftler)
toplam = reduce(lambda acc, x: acc + x, kareler, 0)

print(toplam)  # 220
```
