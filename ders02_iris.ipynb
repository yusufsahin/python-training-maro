```python
import requests
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
out="/content/iris_dataset.csv"

with requests.get(url,stream=True,timeout=60) as r:
  r.raise_for_status()
  with open(out,"wb") as f:
    for chunk in r.iter_content(chunk_size=1024 * 1024):
      if chunk:
        f.write(chunk)
  print("Yazıldı:",out)
```

Daha Temiz kod yazmak için foksiyonlara file işlemleri bölünebilir

```python
from pathlib import Path

def write_text(path: str | Path, text: str, *, append: bool = False) -> None:
    path = Path(path)
    mode = "a" if append else "w"
    path.parent.mkdir(parents=True, exist_ok=True)
    with open(path, mode, encoding="utf-8") as f:
        f.write(text)
def read_text(path: str | Path) -> str:
    path = Path(path)
    with open(path, "r", encoding="utf-8") as f:
        return f.read()
write_text("/content/out/hello.txt", "Selam!\n")
write_text("/content/out/hello.txt", "Ek satır\n", append=True)
print(read_text("/content/out/hello.txt"))

```

```python
import csv

src = "/content/iris_dataset.csv"
dst = "/content/iris_setosa.csv"

TARGET = "setosa"   # filtre kriteri

with open(src, "r", encoding="utf-8", newline="") as fin, \
     open(dst, "w", encoding="utf-8", newline="") as fout:
    reader = csv.reader(fin)
    writer = csv.writer(fout)

    header = next(reader)      # ilk satır başlık
    writer.writerow(header)

    # species sütununun index'ini başlıktan bul
    # iris.csv'de bu sütun genelde "species"
    try:
        species_idx = header.index("species")
    except ValueError:
        raise ValueError(f"'species' sütunu bulunamadı. Başlıklar: {header}")

    kept = 0
    total = 0
    for row in reader:
        total += 1
        # Boş satır / eksik kolon guard
        if not row or len(row) <= species_idx:
            continue

        if row[species_idx] == TARGET:
            writer.writerow(row)
            kept += 1

print("Toplam satır (header hariç):", total)
print("Yazılan (filtrelenen) satır:", kept)
print("Çıktı:", dst)
```
