Encapsulation: private alan + property

Python’da “tam private” yok, ama konvansiyon var:

_x → “iç kullanım”

__x → name-mangling (dışarıdan zor erişilir)


```python
class BankAccount:
  def __init__(self, owner: str, balance: float = 0.0):
    self.owner = owner
    self.__balance = float(balance)

  @property
  def balance(self) -> float:
    return self.__balance

  def deposit(self, amount: float) -> None:
    if amount <= 0:
      raise ValueError("Yatırılan tutar pozitif olmalı.")
    self.__balance += amount

  def withdraw(self, amount: float) -> None:
    if amount <= 0:
      raise ValueError("Çekilen tutar pozitif olmalı.")
    if amount > self.__balance:
      raise ValueError("Yetersiz bakiye.")
    self.__balance -= amount
  def __str__(self):
    return f"Account of {self.owner} with balance {self.__balance}"

acc=BankAccount("John",1000.0)
print(acc)
acc.deposit(500.0)
print(acc)
acc.withdraw(200.0)
print(acc)
```

Inheritance (Kalıtım) + Polymorphism (Çok biçimlilik)

```python
from math import pi
class Shape:
  def area(self)->float:
    raise NotImplementedError

class Rectangle(Shape):
  def __init__(self,width:float,height:float):
    self.width=width
    self.height=height
  def area(self) -> float:
    return self.width * self.height

class Circle(Shape):
  def __init__(self,radius:float):
    self.radius=radius

  def area(self) -> float:
    return pi *self.radius * self.radius

shapes: list[Shape]=[Rectangle(10,20),Circle(5)]
for shape in shapes:
  print(shape.area())

```

classmethod / staticmethod

```python
class Temperature:
  def __init__(self,c:float):
    self.c=c

  @classmethod
  def from_fahrenheit(cls,f:float)->"Temperature":
    return cls((f-32)*5/9)

  @staticmethod
  def is_freezing(c:float)->bool:
    return c<=0

t=Temperature.from_fahrenheit(98.6)
print(t.c)
print(Temperature.is_freezing(t.c))
print(Temperature.is_freezing(-5))
```

Bu metot yeni nesne üretecek mi veya alt sınıfa göre davranmalı mı?
→ @classmethod

Bu metot class-level bir ayarı/konfigürasyonu okuyup değiştirecek mi?
→ @classmethod

Bu metot sadece girdi alıp çıktı üretiyor, self/cls gerekmiyor mu?
→ @staticmethod

classmethod = sınıfı bilir (cls) → üretim / factory / class ayarı

staticmethod = sınıfı bilmez → sadece yardımcı fonksiyon

Composition: “Kalıtımdan çok, parça ekle”

```python
class TodoItem:
  def __init__(self,title):
    self.title=title
    self.done=False

  def mark_done(self):
    self.done=True

class TodoList:
  def __init__(self):
    self.items=[]

  def add(self,title):
    self.items.append(TodoItem(title))

  def complete(self,index):
    self.items[index].mark_done()

  def show(self):
    for i,item in enumerate(self.items):
      status="+" if item.done else "-"
      print(f"{status} {i}.{item.title}")

t = TodoList()
t.add("Study")
t.add("Exercise")
t.complete(0)
t.show()
```
