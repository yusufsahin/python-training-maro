Exceptions

```python
def bol(a,b):
  if b==0:
    raise ValueError("Sifira Bolme")
  return a/b

try:
  print(bol(10,2))
  bol(1,0)
except ValueError as e:
  print("Hata: ",e)
finally:
  print("Bitti")
```

Loglama

```python
import logging

logging.basicConfig(level=logging.INFO)
log=logging.getLogger(__name__)

log.debug("Detay")
log.info("Bilgi")
log.warning("Uyari")
log.error("Hata")
log.critical("Kritik")
```

Unit Testing

```python
import unittest

def topla(a,b):
  return a+b

class TestTopla(unittest.TestCase):
  def test_topla(self):
    self.assertEqual(topla(10,20),30)
unittest.main(argv=[''],exit=False)
```
