```python
from fastapi import FastAPI
import uvicorn
import nest_asyncio
from pydantic import BaseModel

app=FastAPI(title="Colab Fast Api")

class Item(BaseModel):
  name:str
  price:float

@app.get("/")
def root():
  return {"hello":"fast api"}

@app.post("/items")
def create_item(item:Item):
  return {"ok":True,"items":item.model_dump()}

import threading
import nest_asyncio
import uvicorn
from google.colab import output

nest_asyncio.apply()

PORT = 8000

def run():
    uvicorn.run(app, host="0.0.0.0", port=PORT, log_level="info")

threading.Thread(target=run, daemon=True).start()

# Colab içinde yeni sekmede açılabilir bir link verir (senin oturumuna özel)
output.serve_kernel_port_as_window(PORT, path="/")
```
