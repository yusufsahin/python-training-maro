TCP Sunucu ve İstemci

```python
import socket
import threading
import time

HOST='127.0.0.1'
PORT=65432

def basit_sunucu():
  with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR,1)
    s.bind((HOST, PORT))
    s.listen()
    conn, addr = s.accept()
    with conn:
      data=conn.recv(1024).decode('utf-8')
      conn.sendall(('Echo: '+data).encode('utf-8')) # .encode() call was misplaced
def basit_istemci(mesaj):
  with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.connect((HOST, PORT))
    s.sendall(mesaj.encode('utf-8'))
    yanit=s.recv(1024).decode('utf-8')
  return yanit

#Sunucu'yu ayrı thread'de başlat ,sonra istemci ile test et
t=threading.Thread(target=basit_sunucu)
t.deamon=True
t.start()
time.sleep(0.5)

print(basit_istemci('Merhaba'))
```

UDP örneği

```python
UDP_HOST='127.0.0.1'
UDP_PORT=65433

def udp_sunucu():
  with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as s:
    s.bind((UDP_HOST, UDP_PORT))
    data,addr=s.recvfrom(1024)
    s.sendto(b'UDP alindi: ' +data,addr)

def udp_istemci(mesaj):
  with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as s:
    s.connect((UDP_HOST, UDP_PORT))
    s.sendto(mesaj.encode('utf-8'),(UDP_HOST, UDP_PORT))
    yanit,_=s.recvfrom(1024)
  return yanit.decode('utf-8')

t=threading.Thread(target=udp_sunucu)
t.deamon=True
t.start()
time.sleep(0.3)

print(udp_istemci('UDP Test'))
```
